## Introduction
----------------------
The main objective of this project is to create a Credit Model that gives one of three posible scores (Poor, Standard or Good) to the people contained on a database. This database has 100,000 different data of different people and it includes their final score, which will allows to compare and see the accuracy that the model has. The model is based in the FICO model, that has five pillars:

1. Payment history
2. Amount debt
3. Credit history
4. New credit requests
5. Types of credit

To select the variables, it is necessary to see all the database and then pick the data that matter.

The variables selected and the percentage of every pillar is:

| FICO                 |  Own model                                    |  % of final score |
|----------------------|-----------------------------------------------|-------------------|

|  Amount debt         |  Outstanding_Debt                             |  30%              |

|  New credit requests |  Credit_Mix                                   |  15%              |
